## Applications in ML
### Example: PCA using Eigenvalues and Eigenvectors

```python
# Example: PCA using Eigenvalues and Eigenvectors
from sklearn.decomposition import PCA

# Sample data
data = np.array([[2.5, 2.4],
                 [0.5, 0.7],
                 [2.2, 2.9],
                 [1.9, 2.2],
                 [3.1, 3.0],
                 [2.3, 2.7],
                 [2, 1.6],
                 [1, 1.1],
                 [1.5, 1.6],
                 [1.1, 0.9]])

# PCA transformation
pca = PCA(n_components=1)
reduced_data = pca.fit_transform(data)
print("Reduced Data:\n", reduced_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 1: Standardize the data (mean 0, variance 1)
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_standardized = (X - mean) / std

# Step 2: Calculate the covariance matrix
cov_matrix = np.cov(X_standardized, rowvar=False)

# Step 3: Calculate the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Step 4: Sort the eigenvalues and corresponding eigenvectors in descending order
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_idx]
eigenvectors_sorted = eigenvectors[:, sorted_idx]

# Step 5: Choose the top k eigenvectors (2 components in this case)
k = 2
eigenvectors_top_k = eigenvectors_sorted[:, :k]

# Step 6: Project the data onto the top k eigenvectors
X_pca = np.dot(X_standardized, eigenvectors_top_k)

# Visualize the original data (first two features)
plt.figure(figsize=(12, 6))

# Plot original data (2D projection of first two features)
plt.subplot(1, 2, 1)
plt.scatter(X_standardized[:, 0], X_standardized[:, 1], c=y, cmap='viridis')
plt.title('Original Data (Standardized)')
plt.xlabel('Feature 1: Sepal Length')
plt.ylabel('Feature 2: Sepal Width')
plt.colorbar(label='Target Label')

# Plot the data after PCA (2D projection of principal components)
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('Data After PCA (2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target Label')

# Show both plots
plt.tight_layout()
plt.show()

# Print the explained variance ratio
explained_variance = eigenvalues_sorted[:k] / np.sum(eigenvalues_sorted)
print(f'Explained variance ratio: {explained_variance}')
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA (reduce to 2 components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualize original data (4D -> 2D using scatter matrix)
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='viridis')
plt.title('Original Data (2D Projection)')
plt.xlabel('Feature 1: Sepal length')
plt.ylabel('Feature 2: Sepal width')
plt.colorbar(label='Target Label')

# Visualize data after PCA (2D)
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('Data After PCA (2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target Label')

# Show both plots
plt.tight_layout()
plt.show()

# Print the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance ratio: {explained_variance}')
```

## 3. Calculus
### 3.1 Differentiation

```python
"""
Understanding gradients is critical for optimization. Let's visualize derivatives and their role.
"""

# Defining a function
f = lambda x: x**2

def derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

# Visualizing function and derivative
x = np.linspace(-10, 10, 100)
y = f(x)

deriv_values = derivative(f, x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^2')
plt.plot(x, deriv_values, label="f'(x) = 2x", linestyle='--')
plt.title("Function and its Derivative")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data (y = 2x + 1 with some noise)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # Actual model: y = 4 + 3x + noise

# Add a bias term (ones column) to the input data for intercept
X_b = np.c_[np.ones((100, 1)), X]  # Add a column of 1s for the intercept

# Hyperparameters
learning_rate = 0.001
iterations = 2000
m = len(X_b)

# Initialize theta (model parameters)
theta = np.random.randn(2, 1)

# Gradient Descent Function
def gradient_descent(X_b, y, theta, learning_rate, iterations):
    cost_history = []

    for i in range(iterations):
        # Compute the predictions
        predictions = X_b.dot(theta)
        
        # Compute the error
        error = predictions - y
        
        # Calculate the gradient (partial derivatives)
        gradients = (2 / m) * X_b.T.dot(error)
        
        # Update the parameters
        theta -= learning_rate * gradients
        
        # Calculate the cost (Mean Squared Error)
        cost = (1 / m) * np.sum(error ** 2)
        cost_history.append(cost)
        
    return theta, cost_history

# Run gradient descent
theta_optimal, cost_history = gradient_descent(X_b, y, theta, learning_rate, iterations)

# Plot the cost function over iterations
plt.plot(cost_history)
plt.title('Cost Function During Gradient Descent')
plt.xlabel('Iterations')
plt.ylabel('Cost (MSE)')
plt.show()

# Print the final optimal parameters
print(f"Optimal theta values: {theta_optimal}")

# Plot the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, X_b.dot(theta_optimal), color='red', label='Fitted line (After Gradient Descent)')
plt.title('Linear Regression with Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

### 3.3 Integration

```python
# Example: Area Under the ROC Curve in Binary Classification
from sklearn.metrics import auc

# False positive rate (FPR) and true positive rate (TPR)
fpr = np.array([0, 0.1, 0.3, 0.6, 1])
tpr = np.array([0, 0.4, 0.6, 0.8, 1])

# Calculate AUC
roc_auc = auc(fpr, tpr)
print("Area Under the ROC Curve (AUC):", roc_auc)
```

```python
## Integration
"""
Integration evaluates the area under a curve. Let's calculate it for a function.
"""

from scipy.integrate import quad

# Defining the function
def f(x):
    return x**2

# Computing the integral
area, _ = quad(f, 0, 3)

print("Area under f(x) = x^2 from 0 to 3:", area)
```

### 3.4 Optimization Techniques

```python
# Optimization Techniques
"""
Optimization in ML finds the minima of functions. Gradient Descent is a key algorithm.
"""

# Gradient Descent Simulation
f = lambda x: x**2  # Function
f_prime = lambda x: 2*x  # Derivative

def gradient_descent(start, learning_rate, iterations):
    x = start
    history = []
    for _ in range(iterations):
        x -= learning_rate * f_prime(x)
        history.append(x)
    return history

# Running gradient descent
start_point = 10
learning_rate = 0.1
iterations = 20

history = gradient_descent(start_point, learning_rate, iterations)

# Visualizing Gradient Descent
plt.figure(figsize=(10, 6))
x_vals = np.linspace(-10, 10, 100)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x) = x^2')
plt.scatter(history, [f(x) for x in history], color='red', label='Gradient Descent Steps')
plt.title("Gradient Descent")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Example: Logistic Regression Loss Optimization
import matplotlib.pyplot as plt

# Logistic function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Plot the sigmoid function
z = np.linspace(-10, 10, 100)
plt.plot(z, sigmoid(z), label="Sigmoid Function")
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.legend()
plt.grid()
plt.show()
```
