```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def checkDistance(x_cord, y_cord, stars):
    # print(stars)
    for x,y in stars:
        if(math.sqrt((x_cord-x)**2 + (y_cord-y)**2) <= 5000):
            return False
    return True 
```

```python
starsTuples = []
np.random.seed(99)
while (len(starsTuples) < 500):
    x_cord = np.random.randint(-110000, 110000)
    y_cord = np.random.randint(-110000, 110000)
    # print(x_cord, y_cord)
    if(checkDistance(x_cord, y_cord, starsTuples)):
        starsTuples.append((x_cord,y_cord))
```

```python
xpoints, ypoints = zip(*starsTuples)
# print(starsTuples)

# for x,y in starsTuples:
#     plt.plot(x,y)
print(len(starsTuples))
plt.figure(figsize=(20,20))
plt.gca().set_facecolor('black')
plt.scatter(xpoints, ypoints, color="white", s=20, marker="*")
plt.show()
```
