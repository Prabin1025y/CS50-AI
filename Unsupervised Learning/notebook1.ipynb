```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df
```

```python
X = df.iloc[:,[3,4]].values
X
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,14))
dendrogram = sc.dendrogram(sc.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

#Two clusters

hca = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
y_hca = hca.fit_predict(X)

plt.scatter(X[y_hca == 0,0], X[y_hca == 0,1], s=100, c='pink', label='Cluster 1')
plt.scatter(X[y_hca == 1,0], X[y_hca == 1,1], s=100, c='gray', label='Cluster 2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

print(f"Silhoette Score = {silhouette_score(X,y_hca)}")
```

```python
#using 4 clusters
hca = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
y_hca = hca.fit_predict(X)

plt.scatter(X[y_hca == 0,0], X[y_hca == 0,1], s=100, c='pink', label='Cluster 1')
plt.scatter(X[y_hca == 1,0], X[y_hca == 1,1], s=100, c='gray', label='Cluster 2')
plt.scatter(X[y_hca == 2,0], X[y_hca == 2,1], s=100, c='blue', label='Cluster 3')
plt.scatter(X[y_hca == 3,0], X[y_hca == 3,1], s=100, c='cyan', label='Cluster 4')
plt.legend()
plt.show()
print(f"Silhoette Score = {silhouette_score(X,y_hca)}")
```

```python
hca = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hca = hca.fit_predict(X)

plt.scatter(X[y_hca == 0,0], X[y_hca == 0,1], s=100, c='pink', label='Cluster 1')
plt.scatter(X[y_hca == 1,0], X[y_hca == 1,1], s=100, c='gray', label='Cluster 2')
plt.scatter(X[y_hca == 2,0], X[y_hca == 2,1], s=100, c='blue', label='Cluster 3')
plt.scatter(X[y_hca == 3,0], X[y_hca == 3,1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_hca == 4,0], X[y_hca == 4,1], s=100, c='yellow', label='Cluster 5')
plt.legend()
plt.show()
print(f"Silhoette Score = {silhouette_score(X,y_hca)}")
```
